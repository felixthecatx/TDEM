### Calculate Burn Rate per Location

```python
import pandas as pd
pd.set_option('display.max_rows', None)

df = pd.read_csv('D:/TDEM/vaccineprovideraccessibilitydata.csv')

df['Administered'] = df['Total_Shipped'] - df['VACCINES_AVAILABLE']

df['Burn_Rate'] = df['Administered'] / df['Total_Shipped']
df['Burn_Rate'] = df['Burn_Rate'].round(2)

df['Burn_Rate'].sort_values(ascending = False)

# Anomolies in this sense refer to locations which have higher vaccine inventory compared to total shipped. These
# result in negative burn rates per location. 
anomolies = df.loc[df['Burn_Rate'] < 0]
anomolies
```

## Calculate aggregate Burn Rate per County - To be joined with county shapefile

```python

df.drop(df[df['Burn_Rate'] < 0].index, inplace=True)

county_group = df.groupby(['COUNTY']).sum().sort_values(by='Burn_Rate',ascending=False).round(2)

county_group['County_Burn_Rate'] = county_group['Administered'] / county_group['Total_Shipped']

county_group['County_Burn_Rate'] = county_group['County_Burn_Rate'].round(2)

county_group.sort_values(by='County_Burn_Rate',ascending = True).head(6)
county_group.to_csv('D:/TDEM/countyJoinData.csv')
print('complete')
```
